```python
df_unpaid = df[df.Indicator == 'Time spent in unpaid work, by sex'][['Country','Sex','Value','Unit','Indicator']]
```

```python
df_unpaid = df_unpaid[:-2]
```

```python
df_unpaid_men = df_unpaid[df_unpaid['Sex'] == 'Men']
```

```python
df_unpaid_women = df_unpaid[df_unpaid['Sex'] == 'Women']
```

```python
df_unpaid_2 = pd.merge(df_unpaid_men, df_unpaid_women, on = 'Country')
```

```python
df_unpaid_2.rename(columns={'Value_x':'Men',
                          'Value_y':'Women',
                          'Indicator_x':'Indicator'}, 
                 inplace=True)
```

```python
df_unpaid_2 = df_unpaid_2[['Country','Men','Women','Indicator']]
```

```python
df_unpaid_2['gap'] = df_unpaid_2['Women'] - df_unpaid_2['Men']
```

```python
df_unpaid_2 = df_unpaid_2.sort_values(by='Women', ascending=False)
```

```python
df_unpaid_2.head()
```

```python
df_unpaid_2.to_csv('unpaid_3.csv',index=False)
```
