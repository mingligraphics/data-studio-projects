```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import altair as alt
import numpy as np
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
colnames=['Race', '2005', '2015'] 
df = pd.read_csv("race_population.csv", names=colnames, header=None)
```

```python
df.dtypes
```

```python
df['growth'] = 100 * (df['2015'] - df['2005'])/ df['2005']
```

```python
df
```

```python
fig, ax = plt.subplots()
df.sort_values(by='growth').plot(ax = ax, kind='barh',x='Race',y='growth', color='lightgrey', title="Asian Population grows fastest" )
ax.grid('on', axis='x', which='major', linestyle='-', linewidth='0.5', color='lightgrey')
plt.savefig("race_population.pdf",bbox_inches = 'tight')
```

```python
colnames=['Country', '2000', '2015'] 
df2 = pd.read_csv("population.csv", names=colnames, header=None)
```

```python
df2
```

```python
df2['growth'] = 100 * (df2['2015'] - df2['2000'])/ df2['2000']
```

```python
df2 = df2.sort_values(by='growth')
```

```python
df2
```

```python
fig, ax2 = plt.subplots()
df2.plot(ax = ax2, kind='scatter',x=df2['2015'],y=df2['pct'], color='lightgrey' )
ax2.grid('on', axis='x', which='major', linestyle='-', linewidth='0.5', color='lightgrey')
plt.savefig("asian_population_01.pdf",bbox_inches = 'tight')

```
