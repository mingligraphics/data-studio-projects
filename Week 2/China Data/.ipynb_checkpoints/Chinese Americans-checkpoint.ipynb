```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import altair as alt
import numpy as np
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_excel("merged_China.xlsx")
```

```python
df_age = pd.concat([df.head(7).iloc[:,0], df.head(7).iloc[:,1::4]], axis=1)
```

```python
df_age = df_age.T
# df_age.columns = df_age.iloc[0]
```

```python
df_age.columns = df_age.iloc[0]
```

```python
df_age = df_age[1:]
df_age.reset_index()
```

```python
df_age.index
```

```python
ax = df_age.plot(rot=0,figsize=(30, 10))
# ax.set_yticks([])
ax.grid('on', which='major', linestyle='-', linewidth='0.5', color='lightgrey')
ax.set_xticks([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
ax.set_ylim(2000,15000)
plt.savefig("age_merged.pdf")
# ax.set_xticks([])
```

```python
df_state = pd.read_csv("state.csv")
```

```python
df_state
```

```python
df_state_total = pd.concat([df_state.iloc[:,0], df_state.iloc[:,1::3]], axis=1)
```

```python
df_state_total = df_state_total.T
```

```python
df_state_total.columns = df_state_total.iloc[0]
```

```python
df_state_total = df_state_total[1:]
```

```python
df_state_total.reset_index()
```

```python
ax = df_state_total.plot(y='Arizona',figsize=(10, 10))
ax.grid('on', which='major', linestyle='-', linewidth='0.5', color='lightgrey')
# ax.set_xticks([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
ax.set_ylim(0,15000)
plt.savefig("arizona.pdf")
```

```python
melted = df_state_total.melt(id_vars=df_state_total.index, var_name='state', value_name='count')
melted
```

```python
import altair as alt
alt.renderers.enable('notebook')

# https://altair-viz.github.io/gallery/index.html
chart = alt.Chart(melted, width=200, height=200).mark_circle().encode(
    x='month',
    y='crime_type',
    size='crime_count',
    color='crime_type'
)
chart
```
